# **Data Description**

```
ABOUT DATA
----------
A fictional telco company that provided home phone and Internet services to 7043 customers in California in Q3.

COLUMNS EXPLANATION
-------------------
CustomerID: A unique ID that identifies each customer.

Country: The country of the customer’s primary residence.

State: The state of the customer’s primary residence.

Partner: Indicate if the customer has a partner: Yes, No

Dependents: Indicates if the customer lives with any dependents: Yes, No. Dependents could be children, parents, grandparents, etc.

Internet Service: Indicates if the customer subscribes to Internet service with the company: No, DSL, Fiber Optic, Cable.

Online Backup: Indicates if the customer subscribes to an additional online backup service provided by the company: Yes, No

CLTV: Customer Lifetime Value. A predicted CLTV is calculated using corporate formulas and existing data. The higher the value, the more valuable the customer. High value customers should be monitored for churn.
```

### 1.  visualiasi untuk menemukan perbedaan persentase dari `customer yang churn` dan `customer yang tidak churn`

````python
# Menghitung jumlah churn dan tidak churn
churn_counts = df['Churn Label'].value_counts()
````

````python
# Warna yang dipakai
colors = ['#FF6500', '#FFD09B']

# membuat 2 plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# membuat plot ke 1 pie chart
ax1.pie(churn_counts, labels=churn_counts.index, autopct='%1.1f%%', colors=colors)
ax1.set_title('Pie Chart - Churn Distribution')

# plot ke 2 bar chart
sns.barplot(ax=ax2, x=churn_counts.index, y=churn_counts.values, palette=colors)

# menambah nilai tiap batang di sumbu x
for i, v in enumerate(churn_counts.values):
    ax2.text(i, v + 5, str(v), ha='center', va='bottom')

# judul dan label 
ax2.set_title('Bar Chart - Churn vs Non-Churn Count')
fig.suptitle('Comparison of Churn and Non-Churn Customers', fontsize=14)
plt.xlabel('Churn Status')
plt.ylabel('Numbers of Customers')
plt.tight_layout()

plt.show()
````

### Analisis Visualisasi: Perbandingan Pelanggan yang Churn dan Tidak Churn

* **Dominasi Pelanggan yang Tidak Churn:** Grafik pie chart menunjukkan bahwa sebagian besar pelanggan, sekitar 73.5%, memilih untuk tetap menggunakan layanan. Ini mengindikasikan tingkat retensi pelanggan yang cukup baik.
* **Tingkat Churn:** Sebanyak 26.5% pelanggan memutuskan untuk berhenti berlangganan. Meskipun persentase ini lebih kecil dibandingkan dengan pelanggan yang tetap, namun angka ini masih cukup signifikan dan perlu menjadi perhatian.
* **Perbandingan Numerik:** Grafik batang memberikan gambaran yang lebih jelas mengenai jumlah aktual pelanggan yang churn dan tidak churn. Terdapat sekitar 5174 pelanggan yang tidak churn, sedangkan 1869 pelanggan memilih untuk churn.

### **Insight:**

* **Kinerja Retensi yang Baik:** Tingkat retensi pelanggan secara keseluruhan cukup baik, dengan mayoritas pelanggan memilih untuk tetap menggunakan layanan.
* **Potensi Perbaikan:** Meskipun tingkat retensi baik, namun masih ada ruang untuk perbaikan. Persentase 26.5% pelanggan yang churn menunjukkan adanya potensi kehilangan pendapatan dan perlu dilakukan upaya untuk mengurangi angka ini.
* **Fokus pada Pelanggan yang Churn:** Perusahaan perlu melakukan analisis lebih lanjut untuk memahami alasan mengapa pelanggan memilih untuk churn. Faktor-faktor seperti kualitas layanan, harga, persaingan, dan kepuasan pelanggan perlu dipertimbangkan.
* **Strategi Retensi:** Berdasarkan hasil analisis, perusahaan dapat mengembangkan strategi retensi yang lebih efektif. Misalnya, dengan memberikan penawaran khusus kepada pelanggan yang berpotensi churn, meningkatkan kualitas layanan, atau memperbaiki proses penanganan keluhan.


### 2. Buatlah visualisasi untuk melihat hubungan antara `jumlah bulan yang telah dihabiskan customer` di perusahaan pada akhir kuartal dan `total tagihan customer`

````python
# Filter data untuk hanya mengambil 'Tenure Months' pada akhir kuartal (bulan 3, 6, 9, 12, dst.)
kuartal_akhir = df[df['Tenure Months'] % 3 == 0]

# Plotting hubungan antara 'Tenure Months' pada akhir kuartal dan 'Total Charges'
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(data=kuartal_akhir, x='Tenure Months', y='Total Charges', hue='Churn Label', palette='coolwarm', alpha=0.6)

# Tambahkan judul dan label
plt.title('Hubungan Antara Tenure Months pada Akhir Kuartal dan Total Charges', fontsize=14)
plt.xlabel('Tenure Months (Jumlah Bulan)', fontsize=12)
plt.ylabel('Total Charges (Total Tagihan)', fontsize=12)
plt.grid(True)

# Letakkan legenda (legend) di luar grafik
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Churn Label')

# Tampilkan grafik
plt.show()
````

### Analisi Visual Hubungan Antara Tenure Months dan Total Charges

### **Insight Utama:**

1. **Semakin lama pelanggan berlangganan, belum tentu total biaya semakin besar:** Terlihat bahwa ada banyak pelanggan yang telah berlangganan dalam jangka waktu yang lama namun memiliki total biaya yang relatif rendah. Ini mengindikasikan adanya variasi dalam paket layanan atau mungkin adanya promo-promo khusus yang ditawarkan kepada pelanggan lama.
2. **Pelanggan dengan tenure yang lebih pendek cenderung memiliki variasi total biaya yang lebih besar:** Pada rentang tenure yang lebih rendah, kita melihat scatter plot yang lebih menyebar, menunjukkan variasi yang cukup signifikan dalam total biaya. Ini bisa jadi karena adanya perbedaan dalam jenis layanan yang dipilih atau perubahan dalam paket layanan selama masa berlangganan.


### 3. Buatlah visualisasi untuk melihat 5 `kota` teratas yang customernya melakukan churn dan tidak melakukan churn

````python
# filter churn dan nonchurn
churned_customers = df[df['Churn Value'] == 1]
non_churned_customers = df[df['Churn Value'] == 0]

# menghitung jumlah churn dan non churn ditiap ota
churned_city_counts = churned_customers['City'].value_counts()
non_churned_city_counts = non_churned_customers['City'].value_counts()

# menghitung top 5 
top_churned_cities = churned_city_counts.head(5)
top_non_churned_cities = non_churned_city_counts.head(5)
````

````python
top_churned_cities
````

````python
top_non_churned_cities
````

````python
# membuat 2 plot
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

# membuat plot untuk churn
sns.barplot(x=top_churned_cities.index, y=top_churned_cities.values, color='#A5B68D', ax=axes[0])
axes[0].set_title('Top 5 Cities with Highest Churn')
axes[0].set_xlabel('City')
axes[0].set_ylabel('Number of Churned Customers')

for p in axes[0].patches:
    axes[0].annotate(f'{int(p.get_height())}', 
                     (p.get_x() + p.get_width() / 2., p.get_height()), 
                     ha='center', va='bottom')
    
# plot untuk non churn
sns.barplot(x=top_non_churned_cities.index, y=top_non_churned_cities.values, color='#DA8359', ax=axes[1])
axes[1].set_title('Top 5 Cities with Lowest Churn')
axes[1].set_xlabel('City')
axes[1].set_ylabel('Number of Non-Churned Customers')

for p in axes[1].patches:
    axes[1].annotate(f'{int(p.get_height())}', 
                     (p.get_x() + p.get_width() / 2., p.get_height()), 
                     ha='center', va='bottom')

plt.tight_layout()
plt.show()
````

````python
# Menghitung jumlah pelanggan churn dan tidak churn untuk setiap kota
city_churn = df.groupby(['City', 'Churn Label'])['CustomerID'].count().unstack()

# Mengurutkan kota berdasarkan total pelanggan (churn + tidak churn)
city_churn['Total'] = city_churn['Yes'] + city_churn['No']
top_5_cities = city_churn.nlargest(5, 'Total').index
top_5_cities
````

````python
# Menyiapkan data untuk plotting
plot_data = city_churn.loc[top_5_cities, ['Yes', 'No']]

# Membuat plot
plt.figure(figsize=(14, 7))
ax = plot_data.plot(kind='bar', stacked=False, width=0.7, color=['#A5B68D', '#DA8359'])

plt.title('Top 5 Cities with Highest Churn and Lowest Churn', fontsize=14, pad=20)
plt.xlabel('City', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)
plt.legend(['Churn', 'No Churn'], title='Status Churn', loc='upper right')

# Menambahkan label jumlah di luar batang
for container in ax.containers:
    ax.bar_label(container, label_type='edge', padding=2, fontweight='bold')

# Menyesuaikan tampilan
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
sns.despine(left=True, bottom=True)

# Menambahkan padding dan menyesuaikan batas y
plt.tight_layout(pad=3.0)
plt.ylim(0, plt.ylim()[1] * 1.1)  # Menambah 10% ruang di atas batang tertinggi

plt.show()
````

### Analisis Visualisasi: 5 Kota dengan Tingkat Churn Tertinggi dan Terendah

* **Perbandingan Jumlah Pelanggan Churn dan Tidak Churn:** Grafik ini secara jelas menunjukkan perbedaan jumlah pelanggan yang churn dan tidak churn di setiap kota. Kota dengan bar hijau yang lebih tinggi mengindikasikan tingkat churn yang lebih tinggi, sedangkan bar oranye yang lebih tinggi menunjukkan tingkat churn yang lebih rendah.
* **Kota dengan Tingkat Churn Tertinggi:** Berdasarkan grafik, kota dengan tingkat churn tertinggi adalah **Los Angeles**. Kota ini memiliki jumlah pelanggan yang churn jauh lebih banyak dibandingkan dengan kota lainnya dalam daftar ini.
* **Kota dengan Tingkat Churn Terendah:** Sebaliknya, **San Francisco** memiliki tingkat churn terendah di antara 5 kota yang ditampilkan. Jumlah pelanggan yang tidak churn di San Francisco jauh lebih banyak dibandingkan dengan pelanggan yang churn.
* **Perbedaan Tingkat Churn Antar Kota:** Terdapat perbedaan yang signifikan dalam tingkat churn di antara 5 kota ini. Hal ini menunjukkan bahwa faktor-faktor lokal seperti kualitas layanan, persaingan, dan preferensi pelanggan dapat mempengaruhi tingkat churn di masing-masing kota.

### Insight:

* **Variasi Tingkat Churn Antar Kota:** Tingkat churn tidak merata di semua kota. Beberapa kota memiliki masalah churn yang lebih serius dibandingkan dengan kota lainnya.
* **Potensi Masalah di Los Angeles:** Los Angeles perlu menjadi fokus utama dalam upaya mengurangi churn. Perusahaan perlu melakukan analisis lebih lanjut untuk mengidentifikasi penyebab tingginya tingkat churn di kota ini dan mengambil tindakan perbaikan.
* **San Francisco sebagai Contoh Sukses:** San Francisco dapat dijadikan sebagai contoh sukses dalam upaya retensi pelanggan. Perusahaan dapat mempelajari praktik terbaik yang diterapkan di San Francisco dan menerapkannya di kota-kota lain dengan tingkat churn yang tinggi.

### 4. Buatlah visualisasi untuk melihat persebaran data dari `tenure months` untuk masing-masing `churn label`

````python
sns.set(style="whitegrid")
colors = ['#8EACCD', '#D2E0FB']

# membuat plot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Churn Label', y='Tenure Months', data=df, palette=colors)

# custom plotnya
plt.title('Distribution of Tenure Months by Churn Label', fontsize=16)
plt.xlabel('Churn Label', fontsize=14)
plt.ylabel('Tenure Months', fontsize=14)
plt.xticks(rotation=0)
plt.grid(axis='y')

# Show plot
plt.show()
````

### Analisis Visualisasi: Distribusi Tenure Months berdasarkan Churn Label

* **Pelanggan yang Tetap (Churn=No):**
    * **Median:** Durasi berlangganan median pelanggan yang tetap cenderung lebih tinggi dibandingkan dengan pelanggan yang churn. Ini mengindikasikan bahwa pelanggan yang bertahan lebih lama cenderung memiliki durasi berlangganan yang lebih panjang.
    * **Rentang:** Rentang durasi berlangganan juga lebih luas, menunjukkan variasi yang lebih besar dalam durasi berlangganan di antara pelanggan yang tetap. Ini bisa berarti ada kelompok pelanggan yang sangat loyal dengan durasi berlangganan yang sangat panjang.
    * **Outlier:** Terdapat beberapa outlier pada durasi berlangganan yang sangat tinggi pada kelompok pelanggan yang tetap. Ini mungkin menunjukkan adanya segmen pelanggan yang sangat loyal dan bernilai tinggi bagi perusahaan.

* **Pelanggan yang Churn (Churn=Yes):**
    * **Median:** Durasi berlangganan median pelanggan yang churn cenderung lebih rendah. Ini mengindikasikan bahwa pelanggan yang berhenti berlangganan cenderung memiliki durasi berlangganan yang lebih singkat.
    * **Rentang:** Rentang durasi berlangganan lebih sempit dibandingkan dengan pelanggan yang tetap, menunjukkan bahwa durasi berlangganan pelanggan yang churn cenderung lebih terkonsentrasi pada rentang yang lebih rendah.
    * **Outlier:** Jumlah outlier pada durasi berlangganan yang sangat tinggi lebih sedikit dibandingkan dengan kelompok pelanggan yang tetap. Ini mungkin menunjukkan bahwa pelanggan yang churn cenderung memiliki durasi berlangganan yang lebih pendek dan lebih homogen.

### Insight:

* **Hubungan antara Durasi Berlangganan dan Churn:**
    * **Durasi berlangganan yang lebih pendek** cenderung dikaitkan dengan **tingkat churn yang lebih tinggi**. Pelanggan yang baru bergabung cenderung lebih mudah untuk berhenti berlangganan dibandingkan dengan pelanggan yang sudah lama menjadi pelanggan.
    * **Pelanggan dengan durasi berlangganan yang panjang** cenderung lebih **loyal** dan memiliki **nilai yang lebih tinggi** bagi perusahaan.
