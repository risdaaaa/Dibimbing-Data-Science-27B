```python
# Hitung frekuensi untuk setiap jumlah kunjungan
frequency = df['NumWebVisitsMonth'].value_counts().sort_index()
```

```python
# Hitung total data
total_data = len(df)
```

```python
# Hitung PMF (Probabilitas setiap jumlah kunjungan)
pmf = frequency / total_data
```

```python
pmf
```

- Berapa probabilitas seorang pelanggan memiliki 5 kunjungan web
bulanan? Bagaimana probabilitas ini dibandingkan dengan pelanggan
yang memiliki lebih dari 5 kunjungan?

```python
# Probabilitas seorang pelanggan memiliki 5 kunjungan web bulanan
prob_5_visits = pmf.get(5, 0)
print(f"\nProbabilitas seorang pelanggan memiliki 5 kunjungan web bulanan: {prob_5_visits}")

# Probabilitas pelanggan memiliki lebih dari 5 kunjungan web bulanan
prob_more_than_5_visits = pmf[pmf.index > 5].sum()
print(f"Probabilitas seorang pelanggan memiliki lebih dari 5 kunjungan web bulanan: {prob_more_than_5_visits}")
```

```python
# Bandingkan probabilitas
if prob_5_visits > prob_more_than_5_visits:
    print("Peluang pelanggan dengan 5 kunjungan web lebih besar dibandingkan pelanggan dengan lebih dari 5 kunjungan.")
else:
    print("Peluang pelanggan dengan lebih dari 5 kunjungan web lebih besar dibandingkan pelanggan dengan 5 kunjungan.")
```

2. Lakukan pengujian hipotesis untuk membuktikan apakah education
berpengaruh terhadap income atau tidak. Sebagai seorang DS cobalah untuk
menjelaskan Kesimpulan yang Anda dapatkan!

```python
df['Education'].unique()
```

```python
edu_income = df[['Education', 'Income']]
edu_income.head(20)
```

```python
# Hapus nilai yang hilang pada kolom 'Education' dan 'Income'
df_clean = df[['Education', 'Income']].dropna()
```

```python
# Hitung rata-rata pendapatan untuk setiap kategori pendidikan
education_groups_df = df_clean.groupby('Education')['Income'].mean().reset_index()
education_groups_df
```

```python
# Buat grup data berdasarkan kategori pendidikan
grouped_data = [df_clean[df_clean['Education'] == education]['Income'] for education in df_clean['Education'].unique()]

# Lakukan uji ANOVA one way
anova_result = stats.f_oneway(*grouped_data)

# Cetak hasil ANOVA
print(f"F-statistic: {anova_result.statistic}")
print(f"P-value: {anova_result.pvalue}")
```

```python
# Kesimpulan berdasarkan nilai P-value
alpha = 0.05
if anova_result.pvalue <= alpha:
    print("Ada perbedaan signifikan dalam pendapatan berdasarkan tingkat pendidikan (Reject H₀)")
else:
    print("Tidak ada perbedaan signifikan dalam pendapatan berdasarkan tingkat pendidikan (Fail to reject H₀)")
```

### Hipotesis:

- **H₀ (Hipotesis Nol):** Tidak ada perbedaan signifikan dalam pendapatan antara kategori pendidikan yang berbeda. Ini berarti rata-rata pendapatan bagi semua kelompok pendidikan (Graduation, PhD, Master, Basic, 2n Cycle) adalah sama.
- **H₁ (Hipotesis Alternatif):** Ada perbedaan signifikan dalam pendapatan antara kategori pendidikan yang berbeda. Artinya, setidaknya satu kelompok pendidikan memiliki rata-rata pendapatan yang berbeda secara signifikan dari kelompok lain.

### Metode Pengujian:

Untuk menguji hipotesis ini, digunakan **ANOVA (Analysis of Variance)** one way. Metode ini cocok digunakan untuk membandingkan rata-rata pendapatan dari beberapa kelompok (kategori pendidikan dalam hal ini) untuk melihat apakah ada perbedaan yang signifikan.

### P-value:

- **P-value** yang dihasilkan dari uji ANOVA adalah **1.67e-22**, yang sangat kecil dan jauh di bawah tingkat signifikansi standar (α = 0.05).
- P-value ini menunjukkan probabilitas bahwa hasil yang kita amati (atau yang lebih ekstrem) terjadi secara kebetulan jika hipotesis nol benar.

### Significance Value (α):

- **Tingkat signifikansi (α)** yang digunakan adalah 0.05 (5%).
- Ini berarti bahwa jika p-value lebih kecil dari 0.05, kita akan menolak hipotesis nol (H₀) dan menyimpulkan bahwa ada perbedaan signifikan dalam pendapatan berdasarkan kategori pendidikan.

### Kesimpulan:

- Berdasarkan hasil uji ANOVA, p-value yang sangat kecil **(1.67e-22)** berarti kita **menolak hipotesis nol (H₀)**.
- Ini menunjukkan bahwa **ada perbedaan signifikan dalam pendapatan berdasarkan kategori pendidikan**.
